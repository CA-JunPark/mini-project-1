Imports & load dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing

# Load dataset from sklearn
housing = fetch_california_housing(as_frame=True)

df = housing.frame
df.head()
```

Basic information

```python
df.info()
df.describe()
```

Check missing values

```python
df.isnull().sum()
```

Target variable distribution

```python
plt.figure()
plt.hist(df["MedHouseVal"], bins=50)
plt.xlabel("Median House Value")
plt.ylabel("Frequency")
plt.title("Distribution of Median House Value")
plt.show()
```

Correlation heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

Geographical Visualization

```python
longitude = df["Longitude"]
latitude = df["Latitude"]

plt.figure()
scatterplot = sns.scatterplot(data=df, x="Longitude", y="Latitude", hue="MedHouseVal")
plt.grid()
```

# Observations
- `median_income` shows strongest correlation with house value
- Target variable is skewed
- Location features (latitude & longitude) matter. Higher price houses are mostly concentrated along the coastal side of California.
